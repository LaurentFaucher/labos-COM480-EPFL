### Mean temperature over a year

```python
import pandas as pd
```

```python
climate = pd.read_csv('climate_temperature_month_avg.csv')
```

```python
%matplotlib inline
```

```python
climate.plot(x='Year',y='Switzerland')
```

```python
climate_mean = climate.groupby(['Year']).mean()
```

```python
pd.DataFrame.to_csv(climate_mean, 'climate_mean.csv')
```

```python
climate_mean.plot(y='Switzerland')
```
